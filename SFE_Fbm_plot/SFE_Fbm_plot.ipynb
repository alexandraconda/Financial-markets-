```python
import sys
import warnings
from fbm import fbm
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
#n: Number of observation
#H1 and H2: Hurst exponent

n  = 1000
H1 = 0.2
H2 = 0.8
```

```python
z1 = fbm(n=n, hurst=H1, length=1, method='daviesharte')
z2 = fbm(n=n, hurst=H2, length=1, method='daviesharte')
```

```python
f = plt.figure(figsize=(10,6), constrained_layout=True)
spec2 = gridspec.GridSpec(ncols=1, nrows=2, figure=f)
ax1 = f.add_subplot(spec2[0, 0])
ax2 = f.add_subplot(spec2[1, 0])

ax1.plot(z1, color="green")
ax1.set_title('Fractional Brownian motion with H = 0.2', fontsize=20, color="m")
ax1.set_xlabel(' ', fontsize=15, color="m")
ax1.set_ylabel(' ', fontsize=15, color="m")

ax2.plot(z2, color="red")
ax2.set_title('Fractional Brownian motion with H = 0.8', fontsize=20, color="m")
ax2.set_xlabel(' ', fontsize=15, color="m")
ax2.set_ylabel(' ', fontsize=15, color="m")

f.tight_layout()
plt.savefig('Fbmplot.png')
plt.show()
```
