```python
import sys
import math
import random
import warnings
import numpy as np
import matplotlib.pyplot as plt

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
#Tobs: Number of observations for the simulation
#alpha: The rate at which the process mean reverts
#theta: The long run average interest rate
#sigma: The volatility

def SFE_CIRProcess(alpha, theta, sigma, Tobs):
    np.random.seed(123)
    dt = 1./float(Tobs)    
    x = [theta]
    for i in range(Tobs):
        dW = np.random.normal() * np.sqrt(dt)
        dX = alpha*(theta-x[-1])*dt + sigma*np.sqrt(x[-1])*dW
        x.append(x[-1] + dX)
        t = np.array(range(Tobs+1))
        t = t/Tobs
    
    plt.figure(figsize=(8,5))
    plt.plot(t,x, color="green")
    plt.title("Simulated CIR process", fontsize=20, color="m")
    plt.xlabel("t", fontsize=15, color="m")
    plt.ylabel("Values of process", fontsize=15, color="m")
    plt.savefig('CIR Process.png')
    plt.show()
```

```python
SFE_CIRProcess(5, 1, 0.2, 360)
```
