```python
import sys
import math
import random
import warnings
import numpy as np
import matplotlib.pyplot as plt

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
#dt: Delta t (time)
#c: Constant c
#k: Number of trajectories

def SFEWienerProcess(dt, c, k):
    np.random.seed(123)
    k = int(math.floor(k))
    if (dt <= 0 or k <= 0):
        sys.exit("Delta t and number of trajectories must be larger than 0!")
    l = 100
    n = int(math.floor(l/dt))
    t = np.arange(0, n * dt, dt)
    z = np.random.rand(n, k)
    z = np.matrix(z)
    z = 2 * (z > 0.5) - 1 
    dt = math.sqrt(dt)
    z = z * c * dt
    zz = z.cumsum(axis = 0)
    a = np.zeros((1, k))
    x = np.insert(zz, [0], a, axis = 0)
    
    plt.figure(figsize=(8,5))
    plt.plot(x)
    plt.title("Wiener Process", fontsize=20, color="m")
    plt.xlabel("Time t", fontsize=15, color="m")
    plt.ylabel("Values of process", fontsize=15, color="m")
    plt.savefig('Wiener Process.png')
```

```python
SFEWienerProcess(dt = 0.5, c = 1, k = 5) 
```
